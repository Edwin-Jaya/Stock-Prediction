```python
import pandas
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
url = "https://finance.yahoo.com/quote/%5EGSPC/history?p=%5EGSPC"
#url= "https://byjus.com/maths/types-of-data-in-statistics/#:~:text=Types%20of%20Data%20in%20Statistics,%2C%20Ordinal%2C%20Discrete%2C%20Continuous)"
halaman = requests.get(url,headers=headers)
halaman.status_code
```

```python
parser = BeautifulSoup(halaman.text, "html.parser")
parser
```

```python
tabel_setok = parser.find("table",class_="W(100%) M(0)")
tabel_setok
```

```python
data_saham=pandas.DataFrame(columns=['Date','Open','High','Low','Close','Adjclose','Volume'])
```

```python
for baris in tabel_setok.tbody.find_all('tr'):
    kolom = baris.find_all('td')
    date = kolom[0].span.contents[0]
    open_setok = kolom[1].span.contents[0]
    high_setok = kolom[2].span.contents[0]
    low_setok = kolom[3].span.contents[0]
    close_setok = kolom[4].span.contents[0]
    adj_setok = kolom[5].span.contents[0]
    vol_setok = kolom[6].span.contents[0]
    data_saham = data_saham.append({
        'Date':date,
        'Open':open_setok,
        'High':high_setok,
        'Low':low_setok,
        'Close':close_setok,
        'Adjclose':adj_setok,
        'Volume':vol_setok
    }, ignore_index=True)
```

```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(chrome_options=options)
driver.get(url)
delay=20
try:
    halamanMuncul = WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/div[1]/div/div/div/span'))).click()    
    eyer = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="dropdown-menu"]/div/ul[2]/li[2]/button'))).click()    
    time.sleep(5)
    last_scroll_pos=0
    while True:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'body'))).send_keys(Keys.DOWN)
        time.sleep(.01)
        current_scroll_pos=str(driver.execute_script('return window.pageYOffset;'))
        if current_scroll_pos == last_scroll_pos:
            print('scrolling is finished')
            break
        last_scroll_pos=current_scroll_pos
    print("Udah bang, capek")
    parser = BeautifulSoup(driver.page_source, "html.parser")
    tabel_setok = parser.find("table",class_="W(100%) M(0)")
    data_saham_setahun=pandas.DataFrame(columns=['Date','Open','High','Low','Close','Adjclose','Volume'])
    for baris in tabel_setok.tbody.find_all('tr'):
        kolom = baris.find_all('td')
        date = kolom[0].span.contents[0]
        open_setok = kolom[1].span.contents[0]
        high_setok = kolom[2].span.contents[0]
        low_setok = kolom[3].span.contents[0]
        close_setok = kolom[4].span.contents[0]
        adj_setok = kolom[5].span.contents[0]
        vol_setok = kolom[6].span.contents[0]
        data_saham_setahun = data_saham_setahun.append({
            'Date':date,
            'Open':open_setok,
            'High':high_setok,
            'Low':low_setok,
            'Close':close_setok,
            'Adjclose':adj_setok,
            'Volume':vol_setok
        }, ignore_index=True)
    print("Atos, cekap.")
except Exception as e:
    print("Masalah : ",e)
```
